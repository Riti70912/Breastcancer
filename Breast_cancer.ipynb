### Importing the dependencies means imporing the dataset

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Separating the feature and the target

```python
X = breast_cancer.drop(columns='label', axis=1)
Y = breast_cancer['label']
```

```python
print(X)
```

```python
print(Y)
```

Above all we can see that we have separated  the feature and the target variable thourgh its mean values . This we have done to seperate the training and the testing data.

### Spliting the data into taining and the testing data

```python
#separating the training and the testing data and providing the label to them as X & Y
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
#Knowing about the size of the training and the testing data
print(X.shape, X_train.shape, X_test.shape)
```

 Model Training

Logistic Regression

Losgistic regession is the statistical model used for binary classifiaction tasks.It is a supervised learning algorithm , used to predict the probability of the categorical dependent variable based on more than one independent variable .
In logistic regression, the dependent variable is binary and the goal is to find the relationship between the independent variables and the probability of the dependent variable taking a specific value (e.g., 0 or 1). The output of logistic regression is a probability score ranging from 0 to 1.



```python
model=LogisticRegression()
```

```python
#traning the Logistic regression model using the traning data
model.fit(X_train,Y_train)
```

Model Evaluation


Accuracy score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data = ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy)
```
